```python
movies = ["Movie1", "Movie2", "Movie3", "Movie4", "Movie5"]
num_oscars = [5,12,3,6,2]

#default size of bars is 0.8 

xs = [i + 0.5 for i, _ in enumerate(movies)]

plt.bar(xs, num_oscars)

plt.ylabel("Academy Awards")
plt.xlabel("My favorite movies")

plt.xticks([i+0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```

```python
from matplotlib import pyplot as plt
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
gradeMarks = ["A","B","C","D","E","F","G","H","I","J","K"]

# A function that returns the passed in parameter
decile = lambda grade : grade // 10 * 10

# We create a dictionary of keys as the values of the grades list
# And the values being their number of occurences in the grades list
# Helps us to handle duplicates

histogram = Counter(decile(grade) for grade in grades)

# plt.bar plots the given lists
# The first list is the keys of the histogram dictionary
# Like the elements of the grades list but without values 

# The second list are the values of the histogram keys i.e. The number of occurences of the keys in the grades list
# The thirds parameter (8) is to specify the width of the bars in the histogram
plt.bar([x for x in histogram.keys()], histogram.values(), 8)

# With plt.axis we first set the limit of the x-axis(-5 to 105)
# We set the limits of the y-axis too afterwards (0 to 5)
plt.axis([-10,105,0,5])

# plt.xticks helps us set the position of labels and what their values should be
# plt.xticks([i * 10 for i in range(11)],gradeMarks)

plt.xticks([i * 10 for i in range(11)])

print(histogram.keys())

plt.xlabel("Grades")
plt.ylabel("Deciles")

plt.title("Grades and deciles")
plt.show()
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

# zipping returns 
total_error = [x+y for x,y in zip(variance,bias_squared)]

xs = [i for i,_ in enumerate(variance)]

plt.plot(xs,variance, "g-", label="variance")
plt.plot(xs,bias_squared, "b-.", label="bias_squared")
plt.plot(xs, total_error, "b:", label="total_error")

plt.ylabel("")
plt.xlabel("")
plt.title("")

plt.legend(loc = 10)

plt.show()
```
